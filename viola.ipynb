```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.lines import Line2D
```

```python
d2 = []
for i in range(1, 51):
    d2_df = pd.read_csv(f'Data50x50\D{200+i}.POK', sep='\s+')
    d2_df.reset_index(inplace=True)
    d2_df.drop('level_0', axis=1, inplace=True)
    d2.append(d2_df.values)
d2 = np.concatenate(d2)
```

```python
d3 = []
for i in range(1, 51):
    d3_df = pd.read_csv(f'Data50x50\D{300+i}.POK', sep='\s+')
    d3_df.reset_index(inplace=True)
    d3_df.drop('level_0', axis=1, inplace=True)
    d3.append(d3_df.values)
d3 = np.concatenate(d3)
```

```python
fig, axs = plt.subplots(5, 3)

fig.set_size_inches(16, 15)
#fig.tight_layout()

bins = 25
histtype = 'bar'
axs[0][0].hist([d2[:, 0], d3[:, 0]], bins=bins, histtype=histtype)
axs[0][1].hist([d2[:, 1], d3[:, 1]], bins=bins, histtype=histtype)
axs[0][2].hist([d2[:, 2], d3[:, 2]], bins=bins, histtype=histtype)
axs[1][0].hist([d2[:, 3], d3[:, 3]], bins=bins, histtype=histtype)
axs[1][1].hist([d2[:, 4], d3[:, 4]], bins=bins, histtype=histtype)
axs[1][2].hist([d2[:, 5], d3[:, 5]], bins=bins, histtype=histtype)
axs[2][0].hist([d2[:, 6], d3[:, 6]], bins=bins, histtype=histtype)
axs[2][1].hist([d2[:, 7], d3[:, 7]], bins=bins, histtype=histtype)
axs[2][2].hist([d2[:, 8], d3[:, 8]], bins=bins, histtype=histtype)
axs[3][0].hist([d2[:, 9], d3[:, 9]], bins=bins, histtype=histtype)
axs[3][1].hist([d2[:, 10], d3[:, 10]], bins=bins, histtype=histtype)
axs[3][2].hist([d2[:, 11], d3[:, 11]], bins=bins, histtype=histtype)
axs[4][0].hist([d2[:, 12], d3[:, 12]], bins=bins, histtype=histtype)
axs[4][1].hist([d2[:, 13], d3[:, 13]], bins=bins, histtype=histtype)
axs[4][2].hist([d2[:, 14], d3[:, 14]], bins=bins, histtype=histtype)

axs[0][0].set_title('1')
axs[0][1].set_title('2')
axs[0][2].set_title('3')
axs[1][0].set_title('4')
axs[1][1].set_title('5')
axs[1][2].set_title('6')
axs[2][0].set_title('7')
axs[2][1].set_title('8')
axs[2][2].set_title('9')
axs[3][0].set_title('10')
axs[3][1].set_title('11')
axs[3][2].set_title('12')
axs[4][0].set_title('13')
axs[4][1].set_title('14')
axs[4][2].set_title('15')


custom_lines = [Line2D([0], [0], color='C0'),
                Line2D([0], [0], color='C1')]

fig.legend(custom_lines, ['D2', 'D3'], prop={'size': 23})
fig.subplots_adjust(wspace=0.2, hspace=0.4)

plt.show()
```
